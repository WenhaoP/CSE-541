### 1.1

At the stopping time we have:
$$L_{\tau}> 1/\delta$$
$$\Pi^{\tau}_s \frac{p_1(x_s)}{p_0(x_s)}> 1/\delta$$

Since this is a function of random variables:
$$\mathbb{E}_1\bigg[\Pi^{\tau}_s \frac{p_1(x_s)}{p_0(x_s)}\bigg]> 1/\delta$$

Using wald's identity:
$$\mathbb{E}_1\bigg[\Sigma^{\tau}_s \ln{(\frac{p_1(x_s)}{p_0(x_s)})}\bigg]> \ln{1/\delta}$$
$$\mathbb{E}_1[\tau] \mathbb{E}_1\bigg[\ln{(\frac{p_1(x_s)}{p_0(x_s)})}\bigg]> \ln{1/\delta}$$
$$\mathbb{E}_1[\tau] KL(p_1||p_0)> \ln{1/\delta}$$
$$\mathbb{E}_1[\tau]> \frac{\ln{1/\delta}}{ KL(p_1||p_0)}$$

Since $\tau = \min{(t: L_t > 1/\delta})$, we have $L_{t-1} < 1/\delta $.

$$\mathbb{E}[L_{t-1}] < 1/\delta$$

$$\mathbb{E}[\Sigma_t^{\tau-1}\ln{(\frac{p_1(x_t)}{p_0(x_t)})}]< \ln{(1/\delta)}$$

$$(\mathbb{E}[\tau] -1 )\mathbb{E}[\ln{(\frac{p_1(x_t)}{p_0(x_t)})}]<\ln{(1/\delta)}$$

$$(\mathbb{E}[\tau])\mathbb{E}[\ln{(\frac{p_1(x_t)}{p_0(x_t)})}]<\ln{(1/\delta)}+\ln{(\frac{p_1(x_t)}{p_0(x_t)})}$$

By the above:
$$(\mathbb{E}[\tau])\mathbb{E}[\ln{(\frac{p_1(x_t)}{p_0(x_t)})}]<\ln{(1/\delta)}+\ln{(\kappa)}$$

$$\mathbb{E}_1[\tau]< \frac{\ln{\kappa/\delta}}{ KL(p_1||p_0)}$$


### 1.2

Assuming that $p_0,p_1$ are normal , we have:

$$\Pi_i^t\frac{p_1(x_i)}{p_0(x_i)} = \Pi_i^t \exp{(x_i\theta - \theta^2/2)}$$

Distributing the product:
> $$=\exp{(\Sigma_i^t[x_i\theta - \theta^2/2])} = \exp{(\Sigma_i^t[x_i]\theta - t\theta^2/2)} = \exp{(\theta S_t-t\theta^2/2)}= L_t(\theta)$$

To prove if this is Martingale, we must write $\mathbb{E}[L_t]$ in terms of $L_{t-1}$
$$L_t = \exp{(\theta S_t-t\theta^2/2)} = \exp{(\theta S_t)}\exp{(-t\theta^2/2)} = \exp{(\theta X_t)}\exp{(\theta S_{t-1})}\exp{(-(t-1)\theta^2/2)}\exp{(-\theta^2/2)}$$

So the expected value:
$$\mathbb{E}[L_t] = \mathbb{E}[\exp{(\theta X_t)}]\exp{(\theta S_{t-1})}\exp{(-(t-1)\theta^2/2)}\exp{(-\theta^2/2)}$$

Under Hypothesis 0, $X_i$ is 0 mean, and if sub gaussian we have:
$$\mathbb{E}[L_t] \le \exp{(\theta^2/2)}\exp{(\theta S_{t-1})}\exp{(-(t-1)\theta^2/2)}\exp{(- \theta^2/2)}=\exp{(\theta S_{t-1})}\exp{(-(t-1)\theta^2/2)}=L_{t-1}$$

> $\mathbb{E}[L_{t}] \le L_{t-1}$, super martingale

If we mix $L_t$ we have:
$$\mathbb{E}[\bar{L}_t] = \int_\theta\mathbb{E}[L_t(\theta)]h(\theta)d\theta$$

Using the previous step:
$$\mathbb{E}[\bar{L}_t] \le \int_\theta L_{t-1}(\theta)]h(\theta)d\theta$$
$$\mathbb{E}[\bar{L}_t] \le \bar{L}_{t-1}$$

Super.

By maximal innequality we have:
$$\mathbb{P}_0(\bar{L}_t>1/\delta) \le \mathbb{E}_0[\bar{L}_0]\delta$$

At $t=0$ we have $L_t = 0$, so $\mathbb{E}[\bar{L}_t] = 0$, thus:
$$\mathbb{P}_0(\bar{L}_t>1/\delta) \le \delta$$

This applies to any value of t.

Given that $h(\theta) \sim \mathcal{N}$:

$$\bar{L}_t = \sqrt{\frac{\sigma^{-2}}{t+\sigma^{-2}}}\exp{(S_t^2(t+\sigma^{-2})^{-1})}$$

Plugging this into the above expression:

$$\mathbb{P}\bigg(\forall t: t^{-1}|S_t| \ge \sqrt{1+\frac{1}{t\sigma^2}}\sqrt{\frac{2\ln{(1\delta)}+\ln{(t\sigma^2+1)}}{t}}\bigg)\le \delta$$

$S_t$ must be mean 0, so for iid variables $Z_i$ with $\mathbb{E}[Z_i]\neq 0$ we have $S_t = \Sigma_s^t(Z_s-\mathbb{E}[Z])$:

$$\mathbb{P}\bigg(\forall t: t^{-1}|\Sigma_s^t(Z_s-\mathbb{E}[Z])| \ge \sqrt{1+\frac{1}{t\sigma^2}}\sqrt{\frac{2\ln{(1\delta)}+\ln{(t\sigma^2+1)}}{t}}\bigg)\le \delta$$

### 1.3

In order to be successful in identifiying the best arm with $\tau$ pulls each, we have the following condition:

$$\bigcap_i^n\bigg\{|\hat{\theta}_{i,\tau}-\theta_i|\le \epsilon\bigg\}$$

The compliment (bad):

$$\bigcup_i^n\bigg\{|\hat{\theta}_{i,\tau}-\theta_i|\ge \epsilon\bigg\}$$

The difference is 1-sub-gaussian by the fact that they are normal with variance 1, so we apply chernoff and the union bound:
$$\bigcup_i^n\mathbb{P}\bigg\{|\hat{\theta}_{i,\tau}-\theta_i|\ge \epsilon\bigg\} \le \Sigma_i^n 2\exp{(-\tau\epsilon^2/2)}$$

We want this to occur w.p. less than some delta:
$$n 2\exp{(-\tau\epsilon^2/2)} \le \delta$$

And then:
$$\tau \le 2\epsilon^{-2}\ln{(2n/\delta)}$$

In order for the algorithm to be successful, the bound on confidence should be hald the difference between true means eg $\delta/2$:
$$\tau \le 8\Delta^{-2}\ln{(2n/\delta)}$$

### 5.1

```python
delta = 0.05
def bound1(t, delta, sigma2):
    return np.sqrt(2*np.log(2/delta)/t)

def bound2(t, delta, sigma2):
    return np.sqrt(1+1/(t*sigma2))*np.sqrt((2*np.log(1/delta)+np.log(t*sigma2+1))/t)
```

```python
ts = np.linspace(1,10000000, 1000).reshape(-1,1)
sigma2s = np.logspace(-6,0,7).reshape(1,-1)
```

```python
b1 = bound1(ts, delta, sigma2s)
b2 = bound2(ts, delta, sigma2s)
ratio = b2/b1
```

```python
df = pd.DataFrame({'t': ts.reshape(-1)}, index=None)
df[sigma2s.reshape(-1).astype(str)] = ratio
```

```python
df.idxmin()
```

```python
df
```

```python
df_ = df.melt(id_vars=['t'], value_vars=df.columns[1:], value_name='bound ratio', var_name='sigma')
```

```python
df_['b2'] = b2.flatten(order='F')
```

```python
fig, ax = plt.subplots(figsize=(10,5))
sns.lineplot(data=df_, x='t', y='bound ratio', hue='sigma', ax =ax)
ax.set_yscale('log')
ax.set_xscale('log')
```

We see that for this delta, the smaller the sigma, the more time it takes for bound (1) to be at its minimum with respect to bound (3). If the absolute value of the mean is about 0.01, to be sure if it is positive or negative we would need a confidence bound of 0.01. The question is, at what time will we have this confidence? Well we want as tight as possible a bound. Given out confidence bound is 0.01, we want to choose sigma such that the all time bound is closest to the single time bound at that confidence. In this case, that looks to me to happen at about sigma2 = 1e-5, leading to just fewer than 1e5 total samples. See the plot below
